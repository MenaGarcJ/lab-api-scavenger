```python
# Importamos request y definimos la API de GitHub
import requests

url = 'https://api.github.com/'

def ruta(*args):
    return url + '/'.join(args)
```

```python
ruta('users','ironhack-datalabs','repos')
```

```python
repos = requests.get(ruta('users','ironhack-datalabs','repos'))
```
